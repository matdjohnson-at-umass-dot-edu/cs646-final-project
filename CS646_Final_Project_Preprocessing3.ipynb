```python
import os
import re
import gc
from datasets import Dataset, concatenate_datasets, disable_progress_bars
from google.colab import drive

disable_progress_bars()
drive.mount('/content/drive')
```

```python
queries_avg_pattern = "^ms_marco_queries_in_qrel_embs_[0-9]+-[0-9]+_avg$"
queries_avg_concat_dir = "ms_marco_queries_in_qrel_embs_avg"
queries_sum_pattern = "^ms_marco_queries_in_qrel_embs_[0-9]+-[0-9]+_sum$"
queries_sum_concat_dir = "ms_marco_queries_in_qrel_embs_sum"
corpus_not_in_qrel_avg_pattern = "^ms_marco_corpus_not_in_qrel_embs_[0-9]+-[0-9]+_avg$"
corpus_not_in_qrel_avg_concat_dir = "ms_marco_corpus_not_in_qrel_embs_avg"
corpus_not_in_qrel_sum_pattern = "^ms_marco_corpus_not_in_qrel_embs_[0-9]+-[0-9]+_sum$"
corpus_not_in_qrel_sum_concat_dir = "ms_marco_corpus_not_in_qrel_embs_sum"
corpus_in_qrel_avg_pattern = "^ms_marco_corpus_in_qrel_embs_[0-9]+-[0-9]+_avg$"
corpus_in_qrel_avg_concat_dir = "ms_marco_corpus_in_qrel_embs_avg"
corpus_in_qrel_sum_pattern = "^ms_marco_corpus_in_qrel_embs_[0-9]+-[0-9]+_sum$"
corpus_in_qrel_sum_concat_dir = "^ms_marco_corpus_in_qrel_embs_[0-9]+-[0-9]+_sum$"

def concatenate_dataset(dataset_dir_pattern_str, output_dir):
    concatenated_dataset = None
    dataset_path = "/content/drive/MyDrive/CS646-FinalProject/datasets"
    parquet_pattern = re.compile("subset_[0-9]+.parquet")
    dataset_dir_pattern = re.compile(dataset_dir_pattern_str)
    for dataset_dir_file in sorted(os.listdir(dataset_path)):
        print(f"dataset_dir_file: {dataset_dir_file}")
        if dataset_dir_pattern.match(dataset_dir_file) is not None:
            for subset_file in sorted(os.listdir(f"{dataset_path}/{dataset_dir_file}")):
                print(f"subset_file: {subset_file}")
                if parquet_pattern.match(subset_file) is not None:
                    read_dataset = Dataset.from_parquet(f"{dataset_path}/{dataset_dir_file}/{subset_file}")
                    if concatenated_dataset is not None:
                        concatenated_dataset = concatenate_datasets([concatenated_dataset, read_dataset])
                    else:
                        concatenated_dataset = read_dataset
    if not os.path.exists(f"{dataset_path}/{output_dir}"):
        os.makedirs(f"{dataset_path}/{output_dir}")
    print(f"dataset: {concatenated_dataset}")
    concatenated_dataset.to_parquet(f"{dataset_path}/{output_dir}/{output_dir}-concatenated.parquet")
    del concatenated_dataset
    gc.collect()
```

```python
concatenate_dataset(queries_avg_pattern, queries_avg_concat_dir)
```

```python
concatenate_dataset(queries_sum_pattern, queries_sum_concat_dir)
```

```python
concatenate_dataset(corpus_not_in_qrel_avg_pattern, corpus_not_in_qrel_avg_concat_dir)
```

```python
concatenate_dataset(corpus_not_in_qrel_sum_pattern, corpus_not_in_qrel_sum_concat_dir)
```

```python
concatenate_dataset(corpus_in_qrel_avg_pattern, corpus_in_qrel_avg_concat_dir)
```

```python
concatenate_dataset(corpus_in_qrel_sum_pattern, corpus_in_qrel_sum_concat_dir)
```
